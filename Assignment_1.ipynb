```python
df['Country'].head()
```

```python
countries = {}
```

```python
for country_list in df['Country'].str.split(','):
  for country in country_list:
    country_clean = country.strip()

    if country_clean == '':
      continue

    count = countries.get(country_clean, 0)
    countries[country_clean] = count + 1
```

```python
countries.items()
```
